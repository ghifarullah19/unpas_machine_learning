## ✅ **Langkah 8: Persiapan Data untuk Clustering**

Sebelum melakukan clustering, kita perlu mempersiapkan data melalui normalisasi atau standardisasi fitur.

### 🔠 **8.1 Memilih Fitur untuk Clustering**

Pertama, kita perlu memilih fitur yang akan digunakan untuk clustering. Hanya fitur numerik yang bisa digunakan langsung untuk sebagian besar algoritma clustering.

````python
df_clean = df.copy()
````

````python
# Kolom-kolom yang akan digunakan untuk clustering
features_for_clustering = ['Income', 'Kidhome', 'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases',
                           'NumStorePurchases', 'NumWebVisitsMonth', 'TotalSpending', 'NumPurchases']
````

````python
from sklearn.preprocessing import StandardScaler

X = df_clean[features_for_clustering].values

# Menskalakan fitur untuk memastikan semua fitur berkontribusi secara proporsional
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
````

---

## ✅ **Langkah 9: Menentukan Jumlah Cluster Optimal**

Sebelum menerapkan algoritma clustering, penting untuk menentukan jumlah cluster optimal. Ada beberapa metode yang dapat digunakan:

### 📊 **9.1 Metode Elbow**

Metode Elbow menggunakan perubahan inertia (sum of squared distances) terhadap jumlah cluster untuk menemukan jumlah cluster optimal.

````python
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans

# Inisialisasi model KMeans tanpa parameter awal
kmeans = KMeans()

# Inisialisasi visualizer KElbow untuk menentukan jumlah cluster optimal
visualizer = KElbowVisualizer(kmeans, k=(1, 10))

# Fit visualizer engan data untuk menemukan jumlah cluster optimal
visualizer.fit(X_scaled)

# Menampilkan grafik elbow untuk analisis
visualizer.show()
````

* **`KMeans(n_clusters=k)`**: Membuat model KMeans dengan jumlah cluster k.
* **`kmeans.inertia_`**: Jumlah jarak kuadrat dari setiap sampel ke pusat cluster terdekat.


### 📈 **9.2 Metode Silhouette Score**

Silhouette Score mengukur seberapa mirip sebuah objek dengan clusternya sendiri dibandingkan dengan cluster lain. Range nilai dari -1 (clustering terburuk) hingga 1 (clustering terbaik).

````python
from sklearn.metrics import silhouette_score

# Hitung silhouette score untuk berbagai jumlah cluster
silhouette_scores = []
k_range = range(2, 11)  # Silhouette score membutuhkan minimal 2 cluster

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    cluster_labels = kmeans.fit_predict(X_scaled)
    silhouette_avg = silhouette_score(X_scaled, cluster_labels)
    silhouette_scores.append(silhouette_avg)
    print(f"For n_clusters = {k}, the silhouette score is {silhouette_avg:.3f}")

# Visualisasi silhouette score
plt.figure(figsize=(10, 6))
plt.plot(k_range, silhouette_scores, marker='o')
plt.title('Silhouette Score untuk Berbagai Jumlah Cluster')
plt.xlabel('Jumlah Cluster')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()
````

* **`fit_predict()`**: Melatih model dan memprediksi cluster untuk setiap sampel.
* **`silhouette_score()`**: Menghitung silhouette score untuk evaluasi clustering.

---

## ✅ **Langkah 10: Menerapkan Algoritma Clustering**

Setelah menentukan jumlah cluster optimal, kita bisa menerapkan algoritma clustering. Berikut adalah beberapa algoritma clustering yang umum digunakan:

### 🧩 **10.1 K-Means Clustering**

K-Means adalah algoritma clustering yang paling populer dan sederhana. Algoritma ini membagi data menjadi K cluster berdasarkan centroid (pusat cluster).

````python
optimal_k = 3 # Asumsi berdasarkan hasil Elbow dan Silhouette Score
kmeans_final = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
labels = kmeans_final.fit_predict(X_scaled)

# Menambahkan label cluster ke DataFrame yang sudah di-encode
df_clean['cluster'] = labels

print(f"\n--- Analisis Karakteristik Setiap Cluster (dengan {optimal_k} cluster) ---")

# Menghitung rata-rata fitur untuk setiap cluster
cluster_summary = df_clean.groupby('cluster')[features_for_clustering].mean()
print(cluster_summary)
````

* **`KMeans(n_clusters=n_clusters)`**: Membuat model KMeans dengan jumlah cluster yang dipilih.
* **`fit_predict()`**: Melatih model dan memprediksi cluster untuk setiap sampel.
* **`kmeans.cluster_centers_`**: Koordinat centroid untuk setiap cluster.
