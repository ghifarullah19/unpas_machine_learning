## 2. Data Cleaning

```python
dataframe.head()
```

```python
# Mengisi missing value pada numeric columns
numeric_columns = dataframe.select_dtypes(include=np.number)

numeric_columns.head()
```

```python
# Isi missing value
# numeric_columns.dropna(inplace=True)
numeric_columns.fillna(numeric_columns.mean(), inplace=True)

numeric_columns.head()
```

```python
# Mengisi missing value pada categorical columns
categorical_columns = dataframe.select_dtypes(include='object')

categorical_columns.head()
```

```python
# Isi missing value
# categorical_columns.dropna(inplace=True)
categorical_columns.fillna(categorical_columns.mode().iloc[0], inplace=True)

categorical_columns.head()
```

```python
categorical_columns.mode().iloc[0]
```

```python
# Menangani data duplikat
dataframe.drop_duplicates(inplace=True)

dataframe.info()
```

```python
# Menyatukan categorical columns dan numeric columns ke dataframe
dataframe = pd.concat([numeric_columns, categorical_columns], axis=1)

dataframe.head()
```

## 3. Feature Engineering

```python
# Encoding categorical columns
print(categorical_columns['tipe_hunian'].value_counts())
print(categorical_columns['jenis_bahan_bakar'].value_counts())
print(categorical_columns['ac_sentral'].value_counts())
```

```python
# Proses encoding
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

for column in categorical_columns:
  dataframe[column] = label_encoder.fit_transform(dataframe[column])

print(dataframe['tipe_hunian'].value_counts())
print(dataframe['jenis_bahan_bakar'].value_counts())
print(dataframe['ac_sentral'].value_counts())
```

```python
# Normalisasi Numeric Columns
from sklearn.preprocessing import MinMaxScaler

min_max_scaler = MinMaxScaler()

for column in numeric_columns:
  dataframe[column] = min_max_scaler.fit_transform(dataframe[[column]])

dataframe.head()
```

## 4. Data Splitting

### 4.1 Regresi

```python
# Membagi data menjadi fitur dan target
from sklearn.model_selection import train_test_split

fitur = dataframe.drop('konsumsi_minyak_pemanas', axis=1)
target = dataframe['konsumsi_minyak_pemanas']

# Membagi data menjadi data latih dan data uji
fitur_latih, fitur_uji, target_latih, target_uji = train_test_split(
    fitur,
    target,
    test_size=0.2,
    random_state=42
)
```

### 4.2 Klasifikasi

```python
# --- Persiapan Data untuk Klasifikasi ---
# Target adalah 'ac_sentral'
fitur_klasifikasi = dataframe.drop(
    ['ac_sentral', 'konsumsi_minyak_pemanas', 'id_pelanggan'],
    axis=1
) # 'konsumsi_minyak_pemanas' adalah target regresi, 'id_pelanggan' tidak relevan

target_klasifikasi = dataframe['ac_sentral'] # Target sudah di-encode

# Membagi data menjadi data latih dan data uji
fitur_latih_klasifikasi, fitur_uji_klasifikasi, target_latih_klasifikasi, target_uji_klasifikasi = train_test_split(
    fitur_klasifikasi,
    target_klasifikasi,
    test_size=0.2,
    random_state=42,
    stratify=target_klasifikasi # Penting untuk klasifikasi, terutama jika kelas tidak seimbang
)
```

### 4.3 Clustering

```python
# --- Persiapan Data untuk Clustering ---
# Pilih fitur yang relevan untuk clustering.
# Kolom 'id_pelanggan', 'konsumsi_minyak_pemanas' (target regresi), dan 'ac_sentral' (target klasifikasi) sebaiknya dihilangkan.
fitur_cluster = dataframe.drop(
    ['id_pelanggan', 'konsumsi_minyak_pemanas', 'ac_sentral'],
    axis=1
)

print("\n--- Data untuk Clustering ---")
print("Bentuk fitur_cluster:", fitur_cluster.shape)
print("Contoh fitur_cluster:")
print(fitur_cluster.head())
```

## 5. Model Training

### 5.1 Regresi

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Inisialisasi model
lr_model = LinearRegression()
rf_regressor_model = RandomForestRegressor(random_state=42)
```

```python
# --- Pelatihan Model Linear Regression ---
print("--- Melatih Linear Regression ---")
lr_model.fit(fitur_latih, target_latih)
print("Linear Regression terlatih.")

# --- Pelatihan Model Random Forest Regressor ---
print("\n--- Melatih Random Forest Regressor ---")
rf_regressor_model.fit(fitur_latih, target_latih)
print("Random Forest Regressor terlatih.")
```

### 5.2 Klasifikasi

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Inisialisasi Model
logreg_model = LogisticRegression(random_state=42, max_iter=200) # Tingkatkan max_iter jika ada warning konvergensi
rf_classifier_model = RandomForestClassifier(random_state=42)
```

```python
# --- Pelatihan Model Logistic Regression ---
print("\n--- Melatih Logistic Regression ---")
logreg_model.fit(fitur_latih_klasifikasi, target_latih_klasifikasi)
print("Logistic Regression terlatih.")

# --- Pelatihan Model Random Forest Classifier ---
print("\n--- Melatih Random Forest Classifier ---")
rf_classifier_model.fit(fitur_latih_klasifikasi, target_latih_klasifikasi)
print("Random Forest Classifier terlatih.")
```

### 5.3 Clustering

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
# Untuk Elbow Method (opsional)
from yellowbrick.cluster import KElbowVisualizer
```

```python
# --- Menentukan Jumlah Cluster Optimal (Elbow Method - Opsional) ---
print("\n--- Mencari Jumlah Cluster Optimal (Elbow Method) ---")
# Inisialisasi model KMeans untuk visualizer
elbow_model = KMeans(random_state=42, n_init='auto') # n_init='auto' untuk menghindari warning

# Inisialisasi visualizer KElbow
# k adalah rentang jumlah cluster yang akan diuji
visualizer = KElbowVisualizer(elbow_model, k=(2,10), metric='silhouette', timings=False, force_model=True)

visualizer.fit(fitur_cluster) # Fit data ke visualizer
visualizer.show() # Tampilkan plot

# Jumlah cluster optimal berdasarkan Silhouette Score tertinggi dari Elbow plot
n_clusters_optimal = visualizer.elbow_value_ if visualizer.elbow_value_ else 3
print(f"Jumlah cluster optimal yang disarankan (atau default): {n_clusters_optimal}")
```

```python
# --- Pelatihan Model KMeans ---
kmeans_model = KMeans(n_clusters=n_clusters_optimal, random_state=42, n_init='auto')
print(f"\n--- Melatih KMeans dengan {n_clusters_optimal} cluster ---")
cluster_labels = kmeans_model.fit_predict(fitur_cluster)
print("Label cluster untuk setiap data poin:", cluster_labels)
```

## 6. Model Evaluation

### 6.1 Regresi

```python
# --- Evaluasi Model ---
# Prediksi pada data uji
target_pred_lr = lr_model.predict(fitur_uji)
target_pred_rf = rf_regressor_model.predict(fitur_uji)
```

```python
print("\n--- Evaluasi Linear Regression ---")
mae_lr = mean_absolute_error(target_uji, target_pred_lr)
mse_lr = mean_squared_error(target_uji, target_pred_lr)
r2_lr = r2_score(target_uji, target_pred_lr)
print(f"Mean Absolute Error (MAE): {mae_lr:.4f}")
print(f"Mean Squared Error (MSE): {mse_lr:.4f}")
print(f"R-squared (R2): {r2_lr:.4f}")

print("\n--- Evaluasi Random Forest Regressor ---")
mae_rf = mean_absolute_error(target_uji, target_pred_rf)
mse_rf = mean_squared_error(target_uji, target_pred_rf)
r2_rf = r2_score(target_uji, target_pred_rf)
print(f"Mean Absolute Error (MAE): {mae_rf:.4f}")
print(f"Mean Squared Error (MSE): {mse_rf:.4f}")
print(f"R-squared (R2): {r2_rf:.4f}")
```

```python
# Visualisasi hasil prediksi (contoh untuk Random Forest)
plt.figure(figsize=(10, 6))
plt.scatter(target_uji, target_pred_rf, alpha=0.7, label='Prediksi vs Aktual (RF)')
plt.plot([min(target_uji), max(target_uji)], [min(target_uji), max(target_uji)], '--r', lw=2, label='Garis Ideal')
plt.xlabel("Nilai Aktual Heating Oil (Scaled)")
plt.ylabel("Nilai Prediksi Heating Oil (Scaled)")
plt.title("Perbandingan Prediksi vs Aktual Heating Oil (Random Forest Regressor)")
plt.legend()
plt.grid(True)
plt.show()
```

### 6.2 Klasifikasi

```python
# --- Evaluasi Model ---
# Prediksi pada data uji
target_pred_logreg = logreg_model.predict(fitur_uji_klasifikasi)
target_pred_rf_klasifikasi = rf_classifier_model.predict(fitur_uji_klasifikasi)
```

```python
print("\n--- Evaluasi Logistic Regression ---")
accuracy_logreg = accuracy_score(target_uji_klasifikasi, target_pred_logreg)
print(f"Accuracy: {accuracy_logreg:.4f}")
print("Classification Report:")
print(classification_report(target_uji_klasifikasi, target_pred_logreg))
print("Confusion Matrix:")
cm_logreg = confusion_matrix(target_uji_klasifikasi, target_pred_logreg)
sns.heatmap(cm_logreg, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix - Logistic Regression')
plt.xlabel('Prediksi')
plt.ylabel('Aktual')
plt.show()
```

```python
print("\n--- Evaluasi Random Forest Classifier ---")
accuracy_rf_klasifikasi = accuracy_score(target_uji_klasifikasi, target_pred_rf_klasifikasi)
print(f"Accuracy: {accuracy_rf_klasifikasi:.4f}")
print("Classification Report:")
print(classification_report(target_uji_klasifikasi, target_pred_rf_klasifikasi))
print("Confusion Matrix:")
cm_rf_klasifikasi = confusion_matrix(target_uji_klasifikasi, target_pred_rf_klasifikasi)
sns.heatmap(cm_rf_klasifikasi, annot=True, fmt='d', cmap='Greens')
plt.title('Confusion Matrix - Random Forest Classifier')
plt.xlabel('Prediksi')
plt.ylabel('Aktual')
plt.show()
```

### 6.3 Clustering

```python
# Visualisasi cluster (jika fitur bisa direduksi menjadi 2D/3D, misal dengan PCA)
# Untuk kesederhanaan, kita akan memvisualisasikan berdasarkan dua fitur pertama.
if fitur_cluster.shape[1] >= 2:
    plt.figure(figsize=(10, 7))
    sns.scatterplot(x=fitur_cluster.iloc[:, 0], y=fitur_cluster.iloc[:, 1], hue=cluster_labels, palette='viridis', s=100, alpha=0.7)
    # Menambahkan centroid cluster ke plot
    centroids = kmeans_model.cluster_centers_
    plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, color='red', label='Centroids')
    plt.title(f'Visualisasi Cluster (Fitur 1 vs Fitur 2) dengan {n_clusters_optimal} Cluster')
    plt.xlabel(f'Fitur: {fitur_cluster.columns[0]}')
    plt.ylabel(f'Fitur: {fitur_cluster.columns[1]}')
    plt.legend()
    plt.grid(True)
    plt.show()
else:
    print("\nTidak cukup fitur untuk visualisasi scatter plot 2D sederhana.")
```
