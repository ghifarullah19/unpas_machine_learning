## 2. Data Cleaning

Setelah mengetahui ada "kotoran" seperti nilai yang hilang pada data, tahap pembersihan data (data cleaning) ini fokus untuk menangani masalah tersebut agar data menjadi lebih siap untuk dianalisis atau digunakan untuk membuat model.

`````python
dataframe.head()
`````

**Penjelasan Lebih Detail:**
- Perintah ini kembali menampilkan lima baris pertama dari tabel data dataframe. Tujuannya adalah untuk melihat kembali kondisi data, terutama kolom-kolom yang diketahui memiliki nilai hilang (```NaN```), sebelum dilakukan proses pembersihan. Ini membantu untuk membandingkan kondisi data "sebelum" dan "sesudah" pembersihan.

`````python
# Mengisi missing value pada numeric columns
numeric_columns = dataframe.select_dtypes(include=np.number)

numeric_columns.head()
`````

**Penjelasan Lebih Detail:**
- Langkah ini bertujuan untuk memisahkan kolom-kolom yang berisi data numerik (angka) dari DataFrame utama. Ini dilakukan karena strategi untuk menangani nilai hilang pada data numerik bisa berbeda dengan data non-numerik (kategorikal/teks).
- ```numeric_columns = dataframe.select_dtypes(include=np.number)```:
    - ```dataframe.select_dtypes()```: Ini adalah fungsi Pandas yang digunakan untuk memilih kolom berdasarkan tipe datanya.
    - ```include=np.number```: Parameter ini menginstruksikan fungsi untuk hanya memilih kolom-kolom yang tipe datanya dianggap sebagai numerik oleh NumPy (misalnya, ```int64``` untuk angka bulat, ```float64``` untuk angka desimal).
    - Hasil dari pemilihan ini (sebuah DataFrame baru yang hanya berisi kolom-kolom numerik) disimpan dalam variabel ```numeric_columns```.
- ```numeric_columns.head()```: Menampilkan lima baris pertama dari DataFrame ```numeric_columns``` yang baru saja dibuat. Ini memungkinkan untuk memeriksa apakah kolom-kolom numerik, termasuk yang memiliki nilai hilang seperti ```heatingoil```, sudah terpilih dengan benar.

**Konteks Lebih Luas:**
- Memisahkan jenis data adalah praktik umum dalam pra-pemrosesan data. Kolom numerik mungkin memerlukan perlakuan seperti penskalaan atau imputasi (pengisian nilai hilang) dengan metode statistik (rata-rata, median), sedangkan kolom teks mungkin memerlukan encoding atau imputasi dengan modus.

`````python
# Isi missing value
# numeric_columns.dropna(inplace=True)
numeric_columns.fillna(numeric_columns.mean(), inplace=True)

numeric_columns.head()
`````

**Penjelasan Lebih Detail:**
- Ini adalah langkah inti untuk menangani nilai-nilai yang hilang (NaN) yang terdapat dalam kolom-kolom numerik yang sudah dipisahkan sebelumnya.
- ```numeric_columns.fillna(numeric_columns.mean(), inplace=True)```:
    - ```numeric_columns.fillna()```: Fungsi ini digunakan untuk "mengisi" semua nilai ```NaN``` dalam DataFrame ```numeric_columns```.
    - ```numeric_columns.mean()```: Bagian ini menghitung nilai rata-rata (mean) untuk setiap kolom numerik dalam ````numeric_columns````. Nilai rata-rata inilah yang akan digunakan sebagai pengganti untuk setiap nilai NaN yang ditemukan di kolom masing-masing. Jadi, jika ada nilai hilang di kolom ```heatingoil```, nilai tersebut akan diganti dengan rata-rata dari semua nilai ```heatingoil``` yang ada dan valid.
    - ```inplace=True```: Parameter ini sangat penting. Jika disetel ```True```, perubahan (pengisian nilai ```NaN```) akan dilakukan secara langsung pada DataFrame ```numeric_columns``` itu sendiri. Jika ```False``` (atau tidak disertakan), fungsi ```fillna()``` akan menghasilkan DataFrame baru yang sudah diisi, dan ```numeric_columns``` yang asli tidak akan berubah.
- ```# numeric_columns.dropna(inplace=True)```: Baris ini adalah komentar, menunjukkan alternatif lain untuk menangani nilai hilang. Fungsi ```.dropna()``` akan menghapus baris (atau kolom, tergantung bagaimana digunakan) yang mengandung setidaknya satu nilai NaN. Metode ini tidak dipilih di sini; sebaliknya, dipilih metode pengisian.
- ```numeric_columns.head()```: Menampilkan lima baris pertama dari ```numeric_columns``` setelah proses pengisian nilai hilang. **Penjelasan Lebih Detail:**

Ini menggunakan variabel ``categorical_columns`` yang merupakan DataFrame berisi kolom-kolom kategorikal setelah nilai hilangnya diisi dengan modus.
- ``categorical_columns['nama_kolom'].value_counts()``:
    - Fungsi ``.value_counts()`` pada sebuah kolom (``Series``) di Pandas akan menghitung berapa kali setiap nilai unik muncul dalam kolom tersebut.
    - Outputnya akan menunjukkan:
        - Untuk ``home_type``: Jumlah kemunculan '``House``', '``Apartment``', dan '``Condo``'. Karena nilai hilang sebelumnya diisi dengan modus ('``House``'), frekuensi '``House``' akan bertambah satu dari hitungan awal.

**Konteks Lebih Luas:**
- Memahami distribusi kategori penting sebelum melakukan encoding (mengubah teks menjadi angka). Jika ada kategori yang sangat jarang, itu mungkin perlu penanganan khusus. Selain itu, ini adalah kesempatan untuk memverifikasi bahwa pengisian nilai hilang dengan modus telah berjalan sesuai harapan.
- Model machine learning pada umumnya tidak bisa memproses data teks secara langsung; mereka membutuhkan input numerik. Oleh karena itu, kolom-kolom kategorikal ini perlu diubah menjadi representasi angka.

`````python
# Proses encoding
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

# categorical_columns['home_type'] = label_encoder.fit_transform(categorical_columns['home_type'])
# categorical_columns['fueloil'] = label_encoder.fit_transform(categorical_columns['fueloil'])
# categorical_columns['centralair'] = label_encoder.fit_transform(categorical_columns['centralair'])

for column in categorical_columns:
  dataframe[column] = label_encoder.fit_transform(dataframe[column])

print(dataframe['home_type'].value_counts())
print(dataframe['fueloil'].value_counts())
print(dataframe['centralair'].value_counts())
`````

**Penjelasan Lebih Detail:**
- Ini adalah langkah untuk mengubah data teks (kategorikal) di kolom-kolom ``home_type``, ``fueloil``, dan ``centralair`` menjadi representasi numerik menggunakan teknik yang disebut Label Encoding.
- ``from sklearn.preprocessing import LabelEncoder``: Mengimpor alat ``LabelEncoder`` dari paket ``sklearn.preprocessing``. Scikit-learn (``sklearn``) adalah library utama untuk machine learning di Python.
- ``label_encoder = LabelEncoder()``: Membuat sebuah objek ``LabelEncoder``. Objek ini yang akan melakukan pekerjaan pengubahan.
- ``for column in categorical_columns:``: Dilakukan perulangan untuk setiap nama kolom yang ada dalam daftar ``categorical_columns`` (yang sebelumnya berisi kolom-kolom kategorikal yang sudah bersih dari DataFrame utama).
- ``dataframe[column] = label_encoder.fit_transform(dataframe[column])``:
    - ``dataframe[column]``: Ini merujuk pada kolom aktual di DataFrame dataframe yang sudah digabungkan.
    - ``label_encoder.fit_transform(...)``: Metode ini melakukan dua hal secara berurutan pada isi kolom tersebut:
        1. ``fit()``: LabelEncoder "mempelajari" semua nilai unik (kategori teks) yang ada dalam kolom tersebut. Misalnya, untuk ``home_type``, ia akan menemukan '``House``', '``Apartment``', '``Condo``'.
        2. ``transform()``: Setelah mengetahui semua kategori unik, ``LabelEncoder`` akan mengganti setiap kategori teks dengan sebuah angka bulat. Biasanya, angka ini dimulai dari 0, 1, 2, dan seterusnya, seringkali berdasarkan urutan abjad atau urutan pertama kali kategori itu ditemui. Misalnya, '``Apartment``' bisa menjadi ``0``, '``Condo``' menjadi ``1``, dan '``House``' menjadi ``2``.
    - Hasil dari pengubahan ini (kolom yang sekarang berisi angka, bukan teks) akan disimpan kembali ke kolom yang sama di dataframe utama.
- ``print(dataframe['nama_kolom'].value_counts())``: Setelah proses encoding, ``.value_counts()`` kembali dipanggil untuk setiap kolom yang telah diubah. Outputnya sekarang akan menunjukkan angka-angka (hasil encoding) beserta frekuensinya, bukan lagi teks kategori. Jumlah frekuensi untuk setiap representasi angka akan sama dengan frekuensi kategori teks aslinya.

**Konteks Lebih Luas:**
- Mengapa Encoding? Sebagian besar algoritma machine learning dirancang untuk bekerja dengan angka. Mereka tidak bisa memahami teks secara langsung. Oleh karena itu, data kategorikal harus diubah menjadi format numerik.
- Label Encoding: Ini adalah salah satu cara paling sederhana untuk melakukan encoding. Cocok untuk fitur kategorikal yang memiliki urutan intrinsik (misalnya, "kecil", "sedang", "besar" bisa diubah menjadi 0, 1, 2). Namun, jika diterapkan pada fitur yang tidak memiliki urutan (seperti home_type atau fueloil), model machine learning mungkin salah mengartikan bahwa angka yang lebih besar berarti kategori yang "lebih penting" atau "lebih besar nilainya", padahal tidak demikian.
- Alternatif: Untuk fitur kategorikal nominal (tanpa urutan), teknik yang sering lebih disukai adalah One-Hot Encoding, yang membuat kolom baru untuk setiap kategori unik, diisi dengan 0 atau 1. Ini menghindari asumsi urutan tetapi dapat memperbanyak jumlah kolom secara signifikan.

`````python
# Normalisasi Numeric Columns
from sklearn.preprocessing import MinMaxScaler

min_max_scaler = MinMaxScaler()

for column in numeric_columns:
  dataframe[column] = min_max_scaler.fit_transform(dataframe[[column]])

dataframe.head()
`````

**Penjelasan Lebih Detail:**
- Langkah ini bertujuan untuk melakukan normalisasi pada semua kolom numerik dalam dataframe. Normalisasi adalah proses mengubah skala nilai-nilai dalam kolom numerik sehingga semuanya berada dalam rentang yang seragam, biasanya antara 0 dan 1.
- ``from sklearn.preprocessing import MinMaxScaler``: Mengimpor alat ``MinMaxScaler`` dari Scikit-learn. MinMaxScaler adalah salah satu jenis teknik normalisasi.
- ``min_max_scaler = MinMaxScaler()``: Membuat objek MinMaxScaler.
- ``for column in numeric_columns:``: Dilakukan perulangan untuk setiap nama kolom yang ada dalam daftar ``numeric_columns`` (yang berisi nama-nama kolom numerik asli seperti ``id``, ``limit``, ``income``, ``heatingoil``, dll.).
- ``dataframe[column] = min_max_scaler.fit_transform(dataframe[[column]])``:
    - ``dataframe[[column]]``: Ini memilih satu kolom dari dataframe utama. Penggunaan kurung siku ganda ``[[...]]`` penting karena ``fit_transform`` dari ``MinMaxScaler`` mengharapkan input berupa data 2D (seperti tabel dengan satu kolom), bukan hanya deretan angka 1D.
    - ``min_max_scaler.fit_transform(...)``: Metode ini melakukan:
        1. ``fit()``: ``MinMaxScaler`` "mempelajari" nilai minimum dan maksimum aktual dari kolom yang sedang diproses.
        2. ``transform()``: Berdasarkan nilai minimum dan maksimum tersebut, setiap nilai dalam kolom diubah menggunakan rumus: nilai_baru = (nilai_lama - nilai_minimum) / (nilai_maksimum - nilai_minimum). Hasilnya, nilai minimum asli akan menjadi 0, nilai maksimum asli akan menjadi 1, dan semua nilai lain akan berada di antara 0 dan 1.
    - Kolom di dataframe utama akan diperbarui dengan nilai-nilai yang sudah dinormalisasi.
- ``dataframe.head()``: Menampilkan lima baris pertama dari dataframe setelah semua kolom numerik dinormalisasi. Akan terlihat bahwa nilai-nilai di kolom seperti ``id``, ``limit``, ``income``, ``avgtemp``, ``heatingoil``, ``numft``, dll., kini berada dalam rentang 0 hingga 1.

**Konteks Lebih Luas:**
- Mengapa Normalisasi Penting?
    - Skala yang Berbeda: Fitur numerik seringkali memiliki satuan dan rentang nilai yang sangat berbeda (misalnya, usia dalam tahun, pendapatan dalam ribuan atau jutaan, suhu dalam derajat). Jika fitur-fitur ini digunakan langsung, fitur dengan rentang nilai yang lebih besar (misalnya, pendapatan) dapat secara tidak adil mendominasi perhitungan dalam beberapa algoritma machine learning dibandingkan fitur dengan rentang nilai yang lebih kecil (misalnya, jumlah anak).
    - Kinerja Algoritma: Banyak algoritma machine learning, terutama yang berbasis jarak (seperti K-Nearest Neighbors, Support Vector Machines) atau yang menggunakan gradient descent untuk optimasi (seperti Regresi Linear, Regresi Logistik, Jaringan Saraf Tiruan), bekerja lebih baik, lebih cepat konvergen, atau memberikan hasil yang lebih stabil jika semua fitur input memiliki skala yang serupa.
- Min-Max Scaling: Ini adalah metode normalisasi yang sederhana dan efektif. Alternatif lain adalah Standardization (Z-score normalization), yang mengubah data sehingga memiliki rata-rata 0 dan standar deviasi 1. Pilihan antara normalisasi dan standardisasi tergantung pada algoritma yang akan digunakan dan asumsi tentang distribusi data.

## 4. Data Splitting

Setelah data dibersihkan dan fitur-fiturnya direkayasa (di-encode dan dinormalisasi), langkah selanjutnya yang sangat penting adalah membagi data menjadi dua bagian utama: satu bagian untuk "melatih" atau "mengajari" model (training set), dan bagian lainnya untuk "menguji" atau "mengevaluasi" seberapa baik model tersebut setelah dilatih (testing set).

### 4.1 Regresi

Dalam konteks regresi, tujuannya adalah untuk memprediksi sebuah nilai angka yang kontinu. Di sini, tampaknya targetnya adalah kolom ``heatingoil`` (jumlah minyak pemanas yang digunakan).

`````python
# Membagi data menjadi fitur dan target
from sklearn.model_selection import train_test_split

fitur = dataframe.drop('heatingoil', axis=1)
target = dataframe['heatingoil']

# Membagi data menjadi data latih dan data uji
fitur_latih, fitur_uji, target_latih, target_uji = train_test_split(
    fitur,
    target,
    test_size=0.2,
    random_state=42
)
`````

**Penjelasan Lebih Detail:**
- ``from sklearn.model_selection import train_test_split``: Mengimpor fungsi ``train_test_split`` dari Scikit-learn. Fungsi ini adalah alat standar untuk membagi dataset.
- ``fitur = dataframe.drop('heatingoil', axis=1)``:
    - Di sini, dibuat sebuah DataFrame baru yang disebut fitur. DataFrame ini akan berisi semua kolom dari dataframe asli kecuali kolom '``heatingoil``'.
    - ``dataframe.drop('heatingoil', axis=1)``: Perintah ini menghapus kolom yang bernama '``heatingoil``'. Parameter ``axis=1`` menunjukkan bahwa yang dihapus adalah kolom (jika ``axis=0``, yang dihapus adalah baris).
    - Kolom-kolom yang tersisa di fitur ini adalah variabel-variabel independen atau input (sering dilambangkan dengan X) yang akan digunakan oleh model untuk membuat prediksi.
- ``target = dataframe['heatingoil']``:
    - Di sini, dibuat sebuah objek ``Series`` baru (Series adalah struktur data satu dimensi di Pandas, seperti satu kolom tabel) yang disebut target. Objek ini hanya berisi nilai-nilai dari kolom '``heatingoil``'.
    - Kolom '``heatingoil``' ini adalah variabel dependen atau output (sering dilambangkan dengan y) yang ingin diprediksi oleh model.
- ``fitur_latih, fitur_uji, target_latih, target_uji = train_test_split(...)``:
    - Ini adalah baris kunci di mana fungsi ``train_test_split`` digunakan untuk memecah data.
    - fitur dan target yang sudah dipisahkan tadi dimasukkan sebagai input ke fungsi ini.
    - ``test_size=0.2``: Parameter ini menentukan proporsi dari keseluruhan dataset yang akan dialokasikan untuk testing set. Nilai 0.2 berarti 20% dari data akan digunakan untuk pengujian, dan sisanya (80%) akan digunakan untuk pelatihan.
    - ``random_state=42``: Ini adalah parameter yang sangat berguna untuk memastikan bahwa pembagian data menjadi set latih dan uji selalu menghasilkan hasil yang sama setiap kali kode dijalankan. Jika ``random_state`` tidak diatur, setiap kali kode dijalankan, data yang masuk ke set latih dan uji bisa berbeda-beda (karena pembagiannya acak), yang membuat hasil eksperimen sulit direproduksi. Angka 42 adalah pilihan umum, tetapi bisa diganti dengan angka bulat lainnya.
    - Fungsi ini akan mengembalikan empat set data:
        - ``fitur_latih``: Bagian fitur (80%) yang akan digunakan untuk melatih model.
        - ``fitur_uji``: Bagian fitur (20%) yang akan digunakan untuk menguji model.
        - ``target_latih``: Bagian target (80%) yang berpasangan dengan ``fitur_latih``.
        - ``target_uji``: Bagian target (20%) yang berpasangan dengan ``fitur_uji``.

**Konteks Lebih Luas:**
- Pentingnya Pemisahan Data: Bayangkan seorang siswa yang akan menghadapi ujian. Jika siswa tersebut hanya menghafal soal-soal yang akan keluar di ujian (data latih = data uji), ia mungkin mendapatkan nilai sempurna, tetapi itu tidak mencerminkan pemahaman sebenarnya terhadap materi. Ia mungkin akan gagal jika diberi soal yang sedikit berbeda. Sama halnya dengan model machine learning. Jika model diuji menggunakan data yang sama persis dengan yang digunakan untuk melatihnya, performanya mungkin terlihat sangat baik, tetapi itu tidak menjamin model tersebut akan bekerja dengan baik pada data baru yang belum pernah ia lihat sebelumnya.
- Generalisasi: Tujuan utama dari model machine learning adalah kemampuannya untuk generalisasi, yaitu membuat prediksi yang akurat pada data baru. Dengan memisahkan data menjadi set latih dan uji, dapat disimulasikan bagaimana model akan berperilaku pada data yang "tidak dikenal". Performa model pada testing set memberikan indikasi yang lebih baik tentang kemampuannya untuk generalisasi.

- Tujuan klasifikasi berbeda dengan regresi. Regresi memprediksi angka (misalnya, berapa banyak minyak pemanas), sedangkan klasifikasi memprediksi kategori (misalnya, apakah ada pendingin sentral 'Ya' atau 'Tidak').
- Meskipun tujuannya berbeda, prinsip pemisahan data menjadi set latih dan uji tetap sama pentingnya untuk evaluasi kinerja model klasifikasi secara objektif dan untuk menilai kemampuannya dalam mengklasifikasikan data baru dengan benar.

Clustering atau pengelompokan adalah teknik unsupervised learning, yang berarti tidak ada kolom target yang ditentukan sebelumnya yang ingin diprediksi. ### 5.1 Regresi

Model akan dilatih untuk mempelajari bagaimana memprediksi nilai numerik dari ``heatingoil`` berdasarkan fitur-fitur lainnya.

`````python
# Import model
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

# Import metrik evaluasi
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Import library visualisasi
import matplotlib.pyplot as plt
import seaborn as sns
`````

**Penjelasan Lebih Detail:**
- Langkah ini adalah mengimpor "resep" atau jenis-jenis model yang akan dicoba, serta alat untuk mengukur seberapa baik "masakan" (prediksi model) nantinya.
- ``from sklearn.linear_model import LinearRegression``:
    - Mengimpor model Regresi Linear. Ini adalah salah satu model prediksi paling dasar. Bayangkan mencoba menggambar garis lurus yang paling pas melewati sekumpulan titik data. Garis ini kemudian digunakan untuk memprediksi nilai baru. Dalam kasus dengan banyak fitur, ini bukan lagi garis, tapi "bidang datar" dalam ruang multidimensi.
- ``from sklearn.ensemble import RandomForestRegressor``:
    - Mengimpor model Random Forest Regressor. Ini adalah model yang lebih canggih. "Forest" (hutan) berarti ia terdiri dari banyak "pohon" keputusan (decision trees). Setiap pohon membuat prediksinya sendiri, dan hasil akhir adalah rata-rata (atau kombinasi lain) dari prediksi semua pohon. Model ini cenderung lebih akurat dan tahan terhadap overfitting (terlalu hafal data latih) dibandingkan satu pohon keputusan tunggal.
- ``from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score``:
    - Mengimpor fungsi-fungsi untuk mengukur kesalahan atau akurasi model regresi:
        - ``mean_absolute_error (MAE)``: Rata-rata dari selisih absolut (nilai mutlak) antara nilai prediksi dan nilai aktual. Memberikan gambaran seberapa besar rata-rata kesalahan prediksi.
        - ``mean_squared_error (MSE)``: Rata-rata dari kuadrat selisih antara nilai prediksi dan nilai aktual. Karena dikuadratkan, kesalahan yang besar akan memberi dampak lebih signifikan pada skor MSE.
        - ``r2_score`` (R-squared atau Koefisien Determinasi): Mengukur seberapa besar proporsi variasi dalam variabel target (misalnya, ``heatingoil``) yang dapat dijelaskan oleh fitur-fitur input melalui model. Nilainya berkisar dari minus tak hingga sampai 1. Semakin mendekati 1, semakin baik modelnya. Nilai 0 berarti model tidak lebih baik dari hanya memprediksi rata-rata target. Nilai negatif berarti model lebih buruk dari itu.
- ``import matplotlib.pyplot as plt`` dan ``import seaborn as sns``:
    - Mengimpor dua paket program (library) populer untuk membuat grafik dan visualisasi data di Python. matplotlib.pyplot (biasa disingkat plt) adalah dasar untuk plot, sementara seaborn (biasa disingkat sns) dibangun di atas matplotlib dan menyediakan antarmuka yang lebih mudah untuk membuat grafik statistik yang menarik secara visual. Visualisasi sering digunakan untuk memahami data awal, hasil model, atau kesalahan prediksi.

- ``lr_model = LinearRegression()``: Membuat sebuah model Regresi Linear baru dan menyimpannya dalam variabel ``lr_model``. Saat ini, model dibuat dengan pengaturan default (bawaan) dari library.
- ``rf_regressor_model = RandomForestRegressor()``: Membuat sebuah model Random Forest Regressor baru dan menyimpannya dalam variabel ``rf_regressor_model``, juga dengan pengaturan default. Pengaturan default ini seringkali sudah cukup baik untuk memulai, tetapi bisa disesuaikan lebih lanjut untuk meningkatkan performa (proses ini disebut hyperparameter tuning).

**Penjelasan Lebih Detail:**
- Ini adalah tahap di mana model benar-benar "belajar" dari data.
- ``model.fit(fitur_latih, target_latih)``: Metode ``.fit()`` adalah perintah standar di Scikit-learn untuk memulai proses pelatihan model
    - ``fitur_latih``: Ini adalah data fitur (variabel input) dari set data latih yang sudah dipersiapkan sebelumnya (80% dari data asli).

    - Untuk ``lr_model`` (Regresi Linear): Algoritma akan mencari parameter (koefisien atau bobot untuk setiap fitur, dan sebuah intersep) yang menghasilkan garis (atau bidang) yang paling "cocok" dengan data latih, yaitu meminimalkan total kesalahan antara prediksi garis dan nilai target_latih yang sebenarnya.
    - Untuk ``rf_regressor_model`` (Random Forest Regressor): Algoritma akan membangun sejumlah besar pohon keputusan. Setiap pohon akan dilatih pada sampel acak dari ``fitur_latih`` dan ``target_latih``. Proses ini melibatkan pencarian aturan-aturan percabangan pada fitur untuk memprediksi ``target_latih``.
- Setelah proses ``.fit()`` selesai, ``lr_model`` dan ``rf_regressor_model`` kini berisi "pengetahuan" yang telah dipelajari dari data latih dan siap digunakan untuk membuat prediksi.

**Konteks Lebih Luas:**
- Pelatihan model adalah inti dari machine learning. Data latih digunakan untuk "mengkalibrasi" parameter internal model sehingga model dapat menangkap pola dan hubungan yang ada dalam data. Kualitas dan kuantitas data latih sangat mempengaruhi seberapa baik model akan belajar.

### 7.1 Hyperparameter Tuning untuk Regresi

Hyperparameter tuning adalah proses mencari kombinasi parameter terbaik yang tidak dipelajari langsung oleh model dari data. Untuk regresi, kita akan mengoptimasi model Random Forest Regressor yang menunjukkan performa lebih baik dibandingkan Linear Regression.

`````python
# Import library untuk hyperparameter tuning
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.model_selection import cross_val_score

# Definisikan parameter grid untuk Random Forest Regressor
rf_regressor_param_grid = {
    'n_estimators': [50, 100, 200],           # Jumlah pohon dalam forest
    'max_depth': [3, 5, 10, None],            # Kedalaman maksimum pohon
    'min_samples_split': [2, 5, 10],          # Minimum sampel untuk split node
    'min_samples_leaf': [1, 2, 4],            # Minimum sampel di leaf node
    'max_features': ['sqrt', 'log2', None],   # Jumlah fitur yang dipertimbangkan untuk setiap split
    'random_state': [42]                      # Untuk konsistensi hasil
}

print("Parameter grid untuk Random Forest Regressor:")
for param, values in rf_regressor_param_grid.items():
    print(f"  {param}: {values}")
`````

`````python
# Inisialisasi Random Forest Regressor untuk tuning
rf_regressor_tuning = RandomForestRegressor()

# Setup GridSearchCV untuk regresi
rf_regressor_grid_search = GridSearchCV(
    estimator=rf_regressor_tuning,
    param_grid=rf_regressor_param_grid,
    cv=3,                           # 3-fold cross validation (mengingat dataset kecil)
    scoring='r2',                   # Metrik evaluasi: R-squared
    n_jobs=-1,                      # Menggunakan semua core CPU
    verbose=1                       # Menampilkan progress
)

print("Memulai hyperparameter tuning untuk Random Forest Regressor...")
rf_regressor_grid_search.fit(fitur_latih, target_latih)

print("\nHyperparameter tuning selesai!")
print("Best parameters:", rf_regressor_grid_search.best_params_)
print("Best cross-validation R2 score:", rf_regressor_grid_search.best_score_)
`````

`````python
# Evaluasi model yang sudah dioptimasi
best_rf_regressor = rf_regressor_grid_search.best_estimator_
best_rf_regressor_predictions = best_rf_regressor.predict(fitur_uji)

print("=== Perbandingan Performa Regresi ===")
print("\nModel Random Forest Original:")
print(f"MAE: {mae_rf_regressor_predictions:.6f}")
print(f"MSE: {mse_rf_regressor_predictions:.6f}")
print(f"R2: {r2_rf_regressor_predictions:.6f}")

print("\nModel Random Forest Setelah Optimasi:")
mae_best_rf_regressor = mean_absolute_error(target_uji, best_rf_regressor_predictions)
mse_best_rf_regressor = mean_squared_error(target_uji, best_rf_regressor_predictions)
r2_best_rf_regressor = r2_score(target_uji, best_rf_regressor_predictions)

print(f"MAE: {mae_best_rf_regressor:.6f}")
print(f"MSE: {mse_best_rf_regressor:.6f}")
print(f"R2: {r2_best_rf_regressor:.6f}")

# Hitung improvement
mae_improvement = ((mae_rf_regressor_predictions - mae_best_rf_regressor) / mae_rf_regressor_predictions) * 100
mse_improvement = ((mse_rf_regressor_predictions - mse_best_rf_regressor) / mse_rf_regressor_predictions) * 100
r2_improvement = ((r2_best_rf_regressor - r2_rf_regressor_predictions) / abs(r2_rf_regressor_predictions)) * 100

print(f"\nImprovement setelah optimasi:")
print(f"MAE improvement: {mae_improvement:.2f}%")
print(f"MSE improvement: {mse_improvement:.2f}%")
print(f"R2 improvement: {r2_improvement:.2f}%")
`````

### 7.3 Optimasi Clustering: Pemilihan K dan Evaluasi Metrik

Untuk clustering, optimasi berfokus pada pemilihan jumlah cluster (K) yang optimal dan evaluasi kualitas clustering menggunakan berbagai metrik. Meskipun kita telah menggunakan Elbow Method dengan silhouette score, mari kita eksplorasi lebih dalam.

`````python
# Evaluasi berbagai jumlah cluster dengan metrik yang berbeda
from sklearn.metrics import calinski_harabasz_score, davies_bouldin_score
import numpy as np

# Range K yang akan dievaluasi
k_range = range(2, 8)
silhouette_scores = []
calinski_scores = []
davies_bouldin_scores = []
inertias = []

print("Evaluasi clustering untuk berbagai nilai K:")
print("K\tSilhouette\tCalinski-Harabasz\tDavies-Bouldin\tInertia")
print("-" * 70)

for k in k_range:
    # Fit KMeans dengan k clusters
    kmeans_temp = KMeans(n_clusters=k, random_state=42, n_init='auto')
    cluster_labels_temp = kmeans_temp.fit_predict(fitur_cluster)

    # Hitung berbagai metrik
    sil_score = silhouette_score(fitur_cluster, cluster_labels_temp)
    cal_score = calinski_harabasz_score(fitur_cluster, cluster_labels_temp)
    db_score = davies_bouldin_score(fitur_cluster, cluster_labels_temp)
    inertia = kmeans_temp.inertia_

    silhouette_scores.append(sil_score)
    calinski_scores.append(cal_score)
    davies_bouldin_scores.append(db_score)
    inertias.append(inertia)

    print(f"{k}\t{sil_score:.4f}\t\t{cal_score:.4f}\t\t\t{db_score:.4f}\t\t{inertia:.4f}")
`````

`````python
# Visualisasi metrik clustering
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Silhouette Score
axes[0, 0].plot(k_range, silhouette_scores, 'bo-', linewidth=2, markersize=8)
axes[0, 0].set_title('Silhouette Score vs Number of Clusters')
axes[0, 0].set_xlabel('Number of Clusters (K)')
axes[0, 0].set_ylabel('Silhouette Score')
axes[0, 0].grid(True, alpha=0.3)

# Calinski-Harabasz Index
axes[0, 1].plot(k_range, calinski_scores, 'ro-', linewidth=2, markersize=8)
axes[0, 1].set_title('Calinski-Harabasz Index vs Number of Clusters')
axes[0, 1].set_xlabel('Number of Clusters (K)')
axes[0, 1].set_ylabel('Calinski-Harabasz Index')
axes[0, 1].grid(True, alpha=0.3)

# Davies-Bouldin Index
axes[1, 0].plot(k_range, davies_bouldin_scores, 'go-', linewidth=2, markersize=8)
axes[1, 0].set_title('Davies-Bouldin Index vs Number of Clusters')
axes[1, 0].set_xlabel('Number of Clusters (K)')
axes[1, 0].set_ylabel('Davies-Bouldin Index')
axes[1, 0].grid(True, alpha=0.3)

# Inertia (Elbow Method)
axes[1, 1].plot(k_range, inertias, 'mo-', linewidth=2, markersize=8)
axes[1, 1].set_title('Inertia (WCSS) vs Number of Clusters')
axes[1, 1].set_xlabel('Number of Clusters (K)')
axes[1, 1].set_ylabel('Inertia')
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Rekomendasikan K optimal berdasarkan berbagai metrik
optimal_k_silhouette = k_range[np.argmax(silhouette_scores)]
optimal_k_calinski = k_range[np.argmax(calinski_scores)]
optimal_k_davies_bouldin = k_range[np.argmin(davies_bouldin_scores)]

print(f"\nRekomendasi K optimal berdasarkan metrik:")
print(f"Silhouette Score: K = {optimal_k_silhouette}")
print(f"Calinski-Harabasz Index: K = {optimal_k_calinski}")
print(f"Davies-Bouldin Index: K = {optimal_k_davies_bouldin}")
`````
